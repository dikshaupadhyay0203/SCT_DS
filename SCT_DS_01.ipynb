```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\archa\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_5830.csv"
df = pd.read_csv(file_path, skiprows=4)

# Clean column names
df.columns = df.columns.str.strip()

#Country Name and 2022 population
pop_2022 = df[['Country Name', '2022']].dropna()

# Sort by population and take top 20
top20 = pop_2022.sort_values(by='2022', ascending=False).head(20)

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(x='2022', y='Country Name', data=top20,color="#FF69B4")
plt.title('Top 20 Countries by Population in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(pop_2022['2022'], bins=30, kde=True, color='#')
plt.title('Histogram of Country Populations in 2022')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```

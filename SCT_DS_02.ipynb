```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\archa\\Downloads\\train.csv")
print(df.head())
```

```python
#Check for missing values
print("Missing Values:", df.isnull().sum())
#Fill missing values
df['Age'].fillna(df['Age'].median())
df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin'].fillna('Unknown')
```

```python
print("Summary Statistics:")
print(df.describe(include='all'))
```

```python
# Survived Count
sns.countplot(data=df, x='Survived')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival by Gender")
plt.show()

# Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], kde=True, color='#ff69b4')
plt.title("Age Distribution of Passengers")
plt.show()

# Survival by Class
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Gender survival rate
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()
print("\nSurvival Rate by Gender:\n", survival_rate_by_gender)

# Class-wise survival
survival_by_class = df.groupby('Pclass')['Survived'].mean()
print("\nSurvival Rate by Class:\n", survival_by_class)
```

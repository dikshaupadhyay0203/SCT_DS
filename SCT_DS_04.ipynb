```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load your CSV file
df = pd.read_csv("C:\\Users\\archa\\Downloads\\traffic_accidents.csv")

# Cleaning columns 
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]
print("Rows & Columns:", df.shape)
print("Column Names:\n", df.columns.tolist())
df.head()
```

```python
#Weather Condition vs Number of Accidents
plt.figure(figsize=(10, 6))
sns.countplot(y='weather_condition', data=df, order=df['weather_condition'].value_counts().index)
plt.title(" Weather Condition vs Accident Count", fontsize=16, color='deeppink')
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.tight_layout()
plt.show()
```

```python
#Road Surface Condition vs Accidents
plt.figure(figsize=(8, 6))
sns.countplot(x='roadway_surface_cond', data=df, order=df['roadway_surface_cond'].value_counts().index)
plt.title(" Road Surface Condition and Accidents", fontsize=16, color='deeppink')
plt.xticks(rotation=45)
plt.xlabel("Road Surface")
plt.ylabel("Accident Count")
plt.show()
```

```python
#Hour of the Day vs Accident Frequency
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='crash_hour', bins=24, kde=True, color='hotpink')
plt.title(" Hour of Day vs Accident Frequency", fontsize=16, color='deeppink')
plt.xlabel("Hour (0â€“23)")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
#Day of Week vs Accidents
plt.figure(figsize=(10, 6))
sns.countplot(x='crash_day_of_week', data=df, order=sorted(df['crash_day_of_week'].dropna().unique()))
plt.title("Day of Week vs Accidents", fontsize=16, color='deeppink')
plt.xlabel("Day of Week (1=Sun ... 7=Sat)")
plt.ylabel("Accident Count")
plt.tight_layout()
plt.show()
```

```python
#Top 10 Contributing Factors
plt.figure(figsize=(12, 6))
top_factors = df['prim_contributory_cause'].value_counts().head(10)
sns.barplot(x=top_factors.values, y=top_factors.index)
plt.title(" Top 10 Contributing Factors", fontsize=16, color='deeppink')
plt.xlabel("Number of Cases")
plt.ylabel("Contributing Factor")
plt.show()
```
